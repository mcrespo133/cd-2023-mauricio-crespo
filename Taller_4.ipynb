```python
import math
from random import random
import matplotlib.pyplot as plt
import numpy as np


def distancia(x, y):
  return math.sqrt(x**2 + y**2)
```

```python
cont = 0
cantPuntos = 2500

i = 0
while i < cantPuntos:
  x = random()
  y = random()
  i = i+1
  if distancia(x, y) <= 1:
      cont += 1
      plt.plot(x, y, 'ro')
  else:
      plt.plot(x, y, 'bo')

print("Puntos dentro de la circunferencia:", cont, "/", cantPuntos)
print("Estimación de pi:", (cont * 4 / (cantPuntos)))

#Grafica Linea de separación
x = np.arange(-1, 1, 0.01)
y = (1 - x**2)**(1/2)
plt.plot(x,y)

#Ajuste Ejes
plt.axis([0, 1.0, 0, 1.0])

plt.show()
```

```python
cont = 0
puntosLado = 50
paso = 1 / (puntosLado - 1)

i = 0
while i <= 1:
  j = 0
  while j <= 1:
    if distancia(i, j) <= 1:
      plt.plot(i, j, 'ro')
      cont += 1
    else:
      plt.plot(i, j, 'bo')
    j+= paso
  i += paso

print("Puntos:", cont)
print("Estimación de pi:", (cont * 4 / (puntosLado**2)))

#Grafica Linea de separación
x = np.arange(-1, 1, 0.01)
y = (1 - x**2)**(1/2)
plt.plot(x,y)

#Ajuste Ejes
plt.axis([0, 1.0, 0, 1.0])

plt.show()
```
